```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data/speedup_asus.csv")
```

```python
#seq = np.array(data["seq"]).astype(int)*0.01
interval = np.array(range(1,9))
seq_0 = int(data["seq"][0])
seq = seq_0/np.array([seq_0 / (i)for i in interval])

thread = seq_0/np.array(data["Thread"]).astype(int)
omp = seq_0/np.array(data["OMP"]).astype(int)
ff = seq_0/np.array(data["ff"]).astype(int)

```

```python
# multiple line plot
plt.title("Speedup with 10000 agents on asus machine")
plt.plot( interval,seq, markerfacecolor='blue', linewidth=3,color='skyblue', label = "perfect speedup")
plt.plot( interval, omp, markerfacecolor='red', linewidth=3,color='red', label = "omp")
plt.plot( interval,ff,  markerfacecolor='black', linewidth=3,color='black', label = "fast flow")
plt.plot(interval,  thread,  color='green', linewidth=3,  label="thread")
plt.legend()
```
